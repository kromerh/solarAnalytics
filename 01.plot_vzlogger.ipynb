# Plot the Zweiwegez√§hler, read from vzloger


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as md
import numpy as np
```

```python
class VzLogger():
    
    def read_data(self, input_path):
        """
        Read the data from a csv file. does some preprocessing
        INPUT:
            - input_path: path to the csv file.
        OUTPUT:
            - pandas dataframe of the data
        """
        data = pd.read_csv(input_path, index_col=0)
        
        # convert time to datetime
        data['time'] = pd.to_datetime(data['time'])
        
        # drop the id column from mongoDB
        data = data.drop('_id', axis=1)

        # convert units from Wh to kWh
        data['180'] = data['180'] / 1000.0
        data['280'] = data['280'] / 1000.0
        
        return data

    def get_data_hourly(self, data):
        """
        Summarizes the data hourly. Finds the closest hourly value and takes this.
        INPUT:
            - data: pandas dataframe with columns 180, 280, and time.
        OUTPUT:
            - pandas dataframe with hourly value of the energy (columns 180, 280).
        """
        range_start = data['time'].iloc[0].round("1H")
        range_end = data['time'].iloc[-1].round("1H")
        # resample 24 hours in steps of 1H
        s = pd.date_range(start=range_start, end=range_end, freq='1H')
        data_hourly = pd.DataFrame(pd.Series(s, name='time'))
        
        # get the 180 and 280 data closest to this timestamp
        def get_closest(dt, data):
            idx = np.argmin(np.abs(data['time']-dt))
            s1 = data['180'].loc[idx]
            s2 = data['280'].loc[idx]
            return s1, s2

        data_hourly['180'], data_hourly['280'] = zip(*data_hourly['time'].apply(lambda x: get_closest(x, data)))
        
        # return the df
        return data_hourly
        
        
    def difference(self, data_hourly, delta_hours):
        """
        Computes a dataframe with the delta in energy. Needs hourly data!
        INPUT:
            - data_hourly: pandas dataframe with hourly data, cols are index, time, 180, and 280
            - delta_hours: how many hours (integer) the difference should be.
        OUTPUT:
            - pandas dataframe with the differences, index is time
        """
        data_hourly = data_hourly.set_index('time')

        data_differences = data_hourly - data_hourly.shift(delta_hours)

        return data_differences
    
    def remove_large_values(self, data_diff, threshold):
        """
        Some values in the difference data are unreasonably large (logger stopped).
        INPUT: 
            - data_diff: pandas dataframe
            - threshold: float which the data_diff should not be larger or smaller (neg values). others will be removed.
        OUTPUT
            - pandas dataframe, cleaned
        """
        data_diff = data_diff[data_diff['180'] < threshold]
        data_diff = data_diff[data_diff['280'] < threshold]        
        data_diff = data_diff[data_diff['180'] > -threshold]    
        data_diff = data_diff[data_diff['280'] > -threshold]                
        
        return data_diff
    
    def get_data_daily(self, data_hourly):
        """
        Gets the hourly data, taking the timestamp at 00:00:00 o clock
        INPUT:
            - data_hourly: pandas dataframe with hourly data
        OUTPUT:
            - pandas dataframe, values closest to 00:00:00 o clock
        """
        return data_hourly[data_hourly['time'].dt.hour == 0]
    
    def get_data_weekly(self, data):
        """
        Summarizes the data weekly. Finds the closest weekly value and takes this.
        INPUT:
            - data: pandas dataframe with columns 180, 280, and time.
        OUTPUT:
            - pandas dataframe with weekly value of the energy (columns 180, 280).
        """
        range_start = data['time'].iloc[0].round("1H")
        range_end = data['time'].iloc[-1].round("1H")
        # resample 24 hours in steps of 1W
        s = pd.date_range(start=range_start, end=range_end, freq='1W')
        data_hourly = pd.DataFrame(pd.Series(s, name='time'))
        
        # get the 180 and 280 data closest to this timestamp
        def get_closest(dt, data):
            idx = np.argmin(np.abs(data['time']-dt))
            s1 = data['180'].loc[idx]
            s2 = data['280'].loc[idx]
            return s1, s2

        data_hourly['180'], data_hourly['280'] = zip(*data_hourly['time'].apply(lambda x: get_closest(x, data)))
        
        # return the df
        return data_hourly
        
```

```python
# path to csv
input_path = '/Users/hkromer/01_Projects/10.SolarAnlage/latest_vzlogger_reading.csv'

# create object
vz = VzLogger()

# read the data
data = vz.read_data(input_path)

# get the data hourly, this takes some 30seconds
data_hourly = vz.get_data_hourly(data)

# differences
delta_hours = 1
data_diff = vz.difference(data_hourly, delta_hours)

# remove large outsiders
threshold = 6000

data_diff_clean = vz.remove_large_values(data_diff, threshold)

# daily data
data_daily = vz.get_data_daily(data_hourly)
data_daily_diff = vz.difference(data_daily, 1)
data_daily_clean = vz.remove_large_values(data_daily_diff, threshold)

# weekly data
data_weekly = vz.get_data_weekly(data)
data_weekly_diff = vz.difference(data_weekly, 1)
data_weekly_diff_clean = vz.remove_large_values(data_weekly_diff, threshold)
```

## Plot whole time data

```python
X = data['time']
Y1 = data['180'].values
Y2 = data['280'].values
xlabel = 'Zeit'
ylabel = 'Gez√§hlte Energie in kWh'
label1 = '1.8.0'
label2 = '2.8.0'

fig, ax = plt.subplots(figsize=(15,6))
sns.set()
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.lineplot(x=X, y=Y1, ax=ax, label=label1)
sns.lineplot(x=X, y=Y2, ax=ax, label=label2)
ax.xaxis.set_major_formatter(md.DateFormatter('%b %Y'))
ax.set_title(ylabel)
ax.set_xlabel(xlabel)
plt.legend(loc='best')
# plt.savefig(f'{path}pressure_whole_day.png', dpi=1200, format='png')
plt.show()
```
