# Model comparison

In this part of the project to predict the photovoltaic production of solar cells on a roof we are comparing the different regression models that we selected based on their performance in terms of r2_score and mean squared error.

Models we considered can be found here:
- Multi Linear Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-01-multilinearregression#code=shown

- Polynomial Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-02-polynomial-regression#code=shown

- Random Forest Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-03-random-forest-regressionhttps://kyso.io/heiko/predicting-solar-panel-output-model-03-random-forest-regression#code=shown

- Support Vector Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-04-support-vector-regression#code=shown

- Decision Tree Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-05-decision-tree-regression#code=shown

- XGBoost Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-06-xgboost-regression#code=shown

- Artificial Neural Network Regression: https://kyso.io/heiko/predicting-solar-panel-output-model-07-ann-regression#code=shown


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# Gather r2_scores and MSE for each model

```python
data = pd.DataFrame({
    'model': ['Linear','RandomForest', 'SVR', 'DecisionTree', 'XGBoost', 'ANN_32neurons_500epochs', 'ANN_100neurons_1000epochs', 'ANN_32neurons_1000epochs', 'ANN_6neurons_1000epochs'],
    'r2_score': [0.7190894719908503,0.8034519994023561,0.7944537346006463, 0.6783149762507574, 0.7852497173370512, 0.7347774390640276, 0.6639340975614128, 0.7128565927564843, 0.7295538005868867],
    'MSE': [54.70055724462856,38.272987609985634,40.02518288147158, 62.64040789443818, 41.81744348763667, 51.645703633338364, 65.440737535159, 55.914260304370025,52.6628813716351]
})

data
```

# Visualize

```python
# find best MSE
best = data[data.loc[:, 'model'] == 'RandomForest']
x_b = best.index.values
y_b = best['MSE'].values

# find best r2_score
best_r2 = data[data.loc[:, 'model'] == 'ANN_100neurons_1000epochs']
x_b_r2 = best_r2.index.values
y_b_r2 = best_r2['r2_score'].values*100
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.plot(data['r2_score'].values * 100, color='darkblue', lw=3)
plt.plot(data['MSE'].values, color='darkorange', lw=3)

# annotate the best in terms of MSE
plt.scatter(x_b, y_b, color='darkorange', s=80)
ax.annotate(f'{y_b[0]:.1f}', xy=(x_b,y_b+3), fontweight='bold', color='darkorange')

# annotate the best in terms of r2_score
plt.scatter(x_b_r2, y_b_r2, color='darkblue', s=80)
ax.annotate(f'{y_b_r2[0]:.1f}%', xy=(x_b_r2-0.2,y_b_r2+3), fontweight='bold', color='darkblue')


ax.annotate('MSE', xy=(8.1,52), fontweight='bold', color='darkorange')
ax.annotate('r2_score [%]', xy=(8.1,72.5), fontweight='bold', color='darkblue')
plt.ylabel('R2 score [%], MSE [-]')
plt.xticks(range(9), data['model'].values, rotation=90)
# ax.set_xticklabels(data['model'].values, rotation=45)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.tight_layout()
plt.savefig('model_comparison.png',dpi=200)
plt.show()
```

# Conclusion

It seems that the RandomForest model compared best in terms of mean squared error, followed by the XGBoost model. 

In the next step, we will consider time series forecasting and make use of the fact that we want to predict into the future, not treating it as solely a regression problem. Yet, these steps outlined in this series of notebooks were important to have a baseline to compare against.
